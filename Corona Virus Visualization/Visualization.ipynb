```python
import numpy as np
import pandas as pd
```

```python
#read in csv data with index and header dropped
df = pd.read_csv('data/corona_time_series_confirmed_cases.csv', header = 1, index_col = 0)
```

```python
#some data cleaning using Pandas methods
df = (df.drop(columns = 'Unnamed: 3') #drop useless column
      .drop(df.iloc[0,:]) #drop an NA row
      .reset_index(drop=True) #reset the index
     )
```

```python
#melt/gather/pivotlong to make a tidy dataframe
df = df.melt(id_vars=df.columns[0:5],var_name = 'dt',value_name='num_new_cases')
```

```python
#convert datetime to datetime format
df['dt'] = pd.to_datetime(df['dt'])
```
